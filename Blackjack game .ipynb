```python
# This cell contains global initializations
import time
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
```

```python
# Creating Card Class
class Card():
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        return f'{self.suit} of {self.rank}'
```

```python
# Creating Deck Class
class Deck():
    def __init__(self):
        self.set_of_cards = []
        for suit in suits:
            for rank in ranks:
                self.set_of_cards.append(Card(suit,rank))
    def remove_one(self):
        return self.set_of_cards.pop(0)
    def shuffle(self):
        random.shuffle(self.set_of_cards)
```

```python
# Creating Player class 
class Player():
    def __init__(self,name,number_of_chips):
        self.name = name
        self.chips = number_of_chips
    def increment_chips_when_blackjack(self,number_of_chips_bet):
        self.chips += (number_of_chips_bet + number_of_chips_bet/2)
    def increment_chips_won_otherwise(self,number_of_chips_bet):
        self.chips += number_of_chips_bet
    def decrement_chips_when_opponent_blackjack(self,number_of_chips_bet):
        self.chips -= (number_of_chips_bet + number_of_chips_bet/2)
    def decrement_chips_opponent_won_otherwise(self,number_of_chips_bet):
        self.chips -= number_of_chips_bet
    def __str__(self):
        return f'Player {self.name} has currently {self.chips} chips to play'
     
```

```python
# Function to continue the game
def want_to_continue():
    option = input('Do you want to play again? (Y/N) ')
    option = option.lower()
    return option == 'n'
```

```python
# Function to take number of chips to play
def take_chips_count(number_of_chips):
    number_of_chips_to_bet = int(input('How many chips would you like to bet? '))
    if number_of_chips_to_bet > number_of_chips:
        print(f'Sorry you don\'t have {number_of_chips_to_bet} number of chips')
        number_of_chips_to_bet = take_chips_count(number_of_chips)
    elif number_of_chips_to_bet <= 1:
        print('You need atleast 2 chips')
        number_of_chips_to_bet = take_chips_count(number_of_chips)
    elif number_of_chips_to_bet % 2 != 0:
        print('You can only bet even number of chips')
        number_of_chips_to_bet = take_chips_count(number_of_chips)
        
    return number_of_chips_to_bet
```

```python
# Function to ask hit or stand
def ask_to_hit_or_stand():
    choice = input("Would you like to Hit or Stand? (H/S) ")
    choice = choice.lower()
    return choice == 'h'
```

```python
# Function to check player win or not
def check_win(player_cards_values_sum):
    if player_cards_values_sum == 21:
        return True
    else:
        return False
```

```python
# Print details of winner
def print_details_of_winner(player):
    if player.name == 'Dealer(Computer)':
        print(f'       {player.name} won. You loose!')
    
    else:
        print(f'       Congratulations {player.name}! You won.')   
```

```python
# Function to print the cards of user and dealer
def print_cards_of_user_and_dealer(user_cards_values_sum,dealer_cards_values_sum,user_cards_set,dealer_cards_set):
    print(f'Your cards were: ')
    for x in user_cards_set:
        print(x)
    print(f'Your cards sum : {user_cards_values_sum}')
    print('----------------------------')
    print('Dealer\'s cards were:' )
    for y in dealer_cards_set:
          print(y)
    print(f'Dealer\'s cards sum : {dealer_cards_values_sum}')
    
```

```python
# Print details of burst
def print_details_of_burst(user,dealer,user_cards_values_sum,dealer_cards_values_sum,user_cards_set,dealer_cards_set,result):
    print('\n')
    if result == 0:
        
        print('---------------------------------------------------------')
        print(f'      {dealer.name} Burst. Congratulations you win!')
        print('---------------------------------------------------------')
    else:
        print('------------------------------------------------------')
        print(f'         {user.name} Burst! You loose.')
        print('------------------------------------------------------')
       
    print_cards_of_user_and_dealer(user_cards_values_sum,dealer_cards_values_sum,user_cards_set,dealer_cards_set)
    print('----------------------------')
```

```python
# Function to check who win
def check_who_win(user,dealer,user_cards_values_sum,dealer_cards_values_sum,user_cards_set,dealer_cards_set):
    print('\n')
    flag_to_check_who_won = 1
    print('--------------------------------------------------')
    if user_cards_values_sum > dealer_cards_values_sum:
        
        print_details_of_winner(user)
        
        
    elif user_cards_values_sum < dealer_cards_values_sum:
        
        flag_to_check_who_won = 0
        print_details_of_winner(dealer)

    else:
        flag_to_check_who_won = -1

    print('--------------------------------------------------')
    print_cards_of_user_and_dealer(user_cards_values_sum,dealer_cards_values_sum,user_cards_set,dealer_cards_set)
    print('----------------------------')
    return flag_to_check_who_won
```
